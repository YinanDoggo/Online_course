Definitions:
* A _quarter_ is a specific three month period, Q1 is January through March, Q2 is April through June, Q3 is July through September, Q4 is October through December.
* A _recession_ is defined as starting with two consecutive quarters of GDP decline, and ending with two consecutive quarters of GDP growth.
* A _recession bottom_ is the quarter within a recession which had the lowest GDP.
* A _university town_ is a city which has a high percentage of university students compared to the total population of the city.

**Hypothesis**: University towns have their mean housing prices less effected by recessions. Run a t-test to compare the ratio of the mean price of houses in university towns the quarter before the recession starts compared to the recession bottom. (`price_ratio=quarter_before_recession/recession_bottom`)

````python
def get_recession_start():
    '''Returns the year and quarter of the recession start time as a 
    string value in a format such as 2005q3'''
    
    '''A recession is defined as starting with two consecutive quarters 
    of GDP decline, and ending with two consecutive quarters of GDP growth.'''
    
    df_q2 = (pd
             .read_excel("gdplev.xls") #read in excel file
             .rename(columns={"Unnamed: 4":"Year", "Unnamed: 5":"GDP"}) #rename the columns
            )

    indexRows = df_q2[df_q2["Year"] == "2000q1"].index.values.astype(int)[0] #get the index of the actual needed rows
    indexCols = df_q2.columns.get_loc("Year") #get index of the actual needed columns

    df_q2 = (df_q2
             .iloc[indexRows:len(df_q2),indexCols:indexCols+2] #slice and obtain the actual needed df
             .reset_index() #reset the index
             .drop("index", axis=1) #drop the old index
            )

    for i in range(len(df_q2)-2):

        if (df_q2.iloc[i+1,1] < df_q2.iloc[i,1]) & (df_q2.iloc[i+2,1] < df_q2.iloc[i+1,1]): #check whether next year decreases and the year after next decreases further
            RecessionStart = df_q2.iloc[i,0] #get the name of the year where the recession begins
            break
    
    return RecessionStart

get_recession_start()
````

````python
def get_recession_bottom():
    '''Returns the year and quarter of the recession bottom time as a 
    string value in a format such as 2005q3'''
    
    df_q2 = (pd
             .read_excel("gdplev.xls") #read in excel file
             .rename(columns={"Unnamed: 4":"Year", "Unnamed: 5":"GDP"}) #rename the columns
            )

    indexRows1 = df_q2[df_q2["Year"] == get_recession_start()].index.values.astype(int)[0] #get the index of the recession starts
    indexRows2 = df_q2[df_q2["Year"] == get_recession_end()].index.values.astype(int)[0] #get the index of the recession ends
    indexCols = df_q2.columns.get_loc("Year") #get index of the actual needed columns

    df_q2 = (df_q2
             .iloc[indexRows1:(indexRows2 + 1),indexCols:indexCols+2] #slice and obtain the actual needed df
             .reset_index() #reset the index
             .drop("index", axis=1) #drop the old index
            )
    RecessionBottom = (df_q2
                       .loc[df_q2["GDP"] == df_q2["GDP"].min(), "Year"] #Return Serie containing "year" where the "GDP" is the minimum
                       .iloc[0] #Extract "Year" from Serie
                      )
    
    return RecessionBottom

get_recession_bottom()
````
