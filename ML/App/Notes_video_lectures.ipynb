## Histogram

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size) # randomly normally distributed variable with mean(loc) 0 and std(scale) 1
    axs[n].hist(sample)
    axs[n].set_title('n={}'.format(sample_size))
```

change bins size

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(2,2,sharex=True)
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size) # randomly normally distributed variable with mean(loc) 0 and std(scale) 1
    axs[n].hist(sample, bins=100) #change bin size with bins=
    axs[n].set_title('n={}'.format(sample_size))
```

```python
import matplotlib.gridspec as gridspec
plt.figure()
gspec = gridspec.GridSpec(3,3)

top_histogram = plt.subplot(gspec[0,1:]) #index starts now with 0
side_histogram = plt.subplot(gspec[1:,0])
lower_right = plt.subplot(gspec[1:,1:])
```

```python
Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)
lower_right.scatter(X,Y)
top_histogram.hist(X, bins=100)
s = side_histogram.hist(Y, bins=100, orientation='horizontal') #rotate the orientation
```

remove y axis from the top right histogram and the y axis from the bottom left grpah since these are probability density histogram, we only care about the relative values. Clear it.

```python
top_histogram.clear()
top_histogram.hist(X, bins=100, normed=True) #normed=True to tell .hist to scale the probability value between 0 and 1
side_histogram.clear()
side_histogram.hist(Y, bins=100, orientation='horizontal', normed=True)

side_histogram.invert_xaxis() #invert x axis
```

set limit to axis to define same scale for x and y axis for the three plots

```python
for ax in [top_histogram, lower_right]:
    ax.set_xlim(0,1)
for ax in [side_histogram, lower_right]:
    ax.set_ylim(-5,5)
```

## Boxplot

```python
import pandas as pd
normal_sample = np.random.normal(loc=0.0, scale=1.0, size=10000)
random_sample = np.random.random(size=10000)
gamma_sample = np.random.gamma(2, size=10000)

df = pd.DataFrame({'normal':normal_sample,
                  'random':random_sample,
                  'gamma':gamma_sample})
```

```python
plt.clf()
_ = plt.boxplot([df['normal'], df['random'], df['gamma']], whis='range')
```

The boxplot of the random variables following gamma distrubtion looks weird. Plot a histogram of it to show the long tail

```python
plt.figure()
_ = plt.hist(df['gamma'], bins=100)
```

```python
import mpl_toolkits.axes_grid1.inset_locator as mpl_il

plt.figure()
plt.boxplot([df['normal'], df['random'], df['gamma']], whis='range')
ax2 = mpl_il.inset_axes(plt.gca(), width='60%', height='40%', loc=2)
ax2.hist(df['gamma'], bins=100)
ax2.margins(x=0.5)
```

# Animations

```python
import matplotlib.animation as animation

n = 100
x = np.random.randn(n)
```

```python
def update(curr):
    if curr == n:
        a.event_source.stop()
    plt.cla()
    bins = np.arange(-4,4,0.5)
    plt.hist(x[:curr],bins=bins)
    plt.axis([-4,4,0,30])
    plt.gca().set_title('Sampling the Normal Distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlable('Value')
    plt.annotate('n = {}'.format(curr),[3,27]) #to show in the animation which curr is currently sampled
```

```python
fig = plt.figure()
a = animation.FuncAnimation(fig, update, interval=100) #interval=100 update every 100 milisecond
```

# Interactivity

```python
plt.figure()
data = np.random.rand(10)
plt.plot(data)

def onclick(event):
    plt.cla()
    plt.plot(data)
    plt.gca().set_title('Event at pixels {},{} {}and data {},{}'.format(event.x,
                                                                       event.y,
                                                                       '\n',
                                                                       event.xdata,
                                                                       event.ydata))
    
plt.gcf().canvas.mpl_connect('botton_press_event',onclick) #connect to event listener through wiring up
```

```python
from random import shuffle
import pandas as pd
origins = ['China','Brazil','India','USA','Canada','UK','Germany','Iraq','Chile','Mexico']

shuffle(origins)

df = pd.DataFrame({'height':np.random.rand(10),
                  'weight':np.random.rand(10),
                   'origin':origins})

df
```

```python
plt.figure()
plt.scatter(df['height'],df['weight'],picker=5) #picker defines the pixels near the data point to activate the animation
plt.gca().set_ylabel('Weight')
plt.gca().set_xlabel('Height')
```

```python
def onpick(event):
    origin = df.iloc[event.ind[0]]['origin']
    plt.gca().set_title('Selected item came from {}'.format(origin))
plt.gcf().canvas.mpl_connect('pick_event',onpick)
```
